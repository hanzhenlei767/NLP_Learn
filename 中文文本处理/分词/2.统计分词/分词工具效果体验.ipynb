# 读取待分词语料

```python
# load test sentences
with open(file="./sentences", encoding="utf-8",) as fr:
    lines = fr.readlines()
sentences = [line[:-1] for line in lines]  #删除换行符
```

```python
#查看前5条内容
sentences[:5]
```

# 1.jieba

```python
import jieba
```

```python
for i, sent in enumerate(sentences):
    print("=="*20+"sentence {}".format(i)+"=="*20)
    print("      raw_sentence:\t", sent)
    jieba_seg_result = jieba.lcut(sent)
    print("  jieba_seg_result:\t", jieba_seg_result)
    
```

# 2.thulac

```python
import thulac
thulac1 = thulac.thulac(seg_only=True)
```

```python
for i, sent in enumerate(sentences):
    print("=="*20+"sentence {}".format(i)+"=="*20)
    print("      raw_sentence:\t", sent)
    result = thulac1.cut(sent)
    thulac_seg_result = [i[0] for i in result]
    print(" thulac_seg_result:\t", thulac_seg_result)
    
```

# 3.snownlp

```python
from snownlp import SnowNLP
```

```python
for i, sent in enumerate(sentences):
    print("=="*20+"sentence {}".format(i)+"=="*20)
    print("      raw_sentence:\t", sent)
    snownlp_seg_result = SnowNLP(sent).words
    print("snownlp_seg_result:\t", snownlp_seg_result)
```

# 4.pkuseg

```python
import pkuseg
seg = pkuseg.pkuseg()
```

```python
for i, sent in enumerate(sentences):
    print("=="*20+"sentence {}".format(i)+"=="*20)
    print("      raw_sentence:\t", sent)
    pku_seg_result = seg.cut(txt=sent)
    print("    pku_seg_result:\t", pku_seg_result)  
```

# 5.pynlpir

```python
import pynlpir
```

```python
pynlpir.open()
```

```python
for i, sent in enumerate(sentences):
    print("=="*20+"sentence {}".format(i)+"=="*20)
    print("      raw_sentence:\t", sent)
    pynlpir_seg_result = pynlpir.segment(sent,pos_tagging=False)
    print("pynlpir_seg_result:\t", pynlpir_seg_result)
```
